Importing packages


```python
import json
import config
import ccxt
import time
```

AUTHENTICATING WITH THE BINANCE CLIENT

```python

exchange = ccxt.binance({
    'apiKey': config.BINANCE_APIKEY,
    'secret' : config.BINANCE_SECRET,
})

exchange.set_sandbox_mode(True) #remove this line for testing on the mainnet, included for testnet environment
symbol = 'BTC/USDT'
interval = '1m' #can be changed to: 5m, 15m, 30m, 1h
interval_in_minutes = 5 #60 for 1h chart, need to be changed according to above variable

amount=0.001
```

```python

def cal_ema(period,previous, closing):
    multiplier = 2/(period+1)
    ema = (closing *multiplier) + ((1-multiplier)* previous)
    return ema
    
```

```python
x = exchange.fetch_ohlcv(symbol,interval)
sma12_data = x[-12:]
sma26_data = x[-26:]
count = 0
sum = 0
for i in sma12_data:
    sum += i[4] #i[4] has the closing value

    count += 1
sma12 = sum/count
count = 0
sum = 0
for i in sma26_data:
    sum += i[4] #i[4] has the closing value

    count += 1
sma26 = sum/count
x = x[-1][4]

ema12 = cal_ema(12, sma12,x)
ema26 = cal_ema(26,sma26,x)
while True:
    x = exchange.fetch_ohlcv(symbol,interval)
    x = x[-1][4]    
    ema_12 = cal_ema(12, ema12,x)
    ema_26 = cal_ema(26,ema26,x)
    print(ema_12)
    print(ema_26)
    with open("flag.json", "r") as file:
        json_object = json.load(file)
        flag = json_object["bema"]
        #print(flag)
    
    if (ema_12 > ema_26) and flag=="false":
        x = exchange.create_order(symbol=symbol,side = "buy", type="market", amount=amount)
        print(x)
        file.close()
        json_object["bema"] = "true"
        file = open("flag.json", "w")
        json.dump(json_object, file)
        file.close()

    else:
        if flag=="true" and (ema_12 < ema_26):
            x = exchange.create_order(symbol=symbol,side = "sell", type="market", amount=amount)
            print(x)
            json_object["bema"] = "false"
            file = open("flag.json", "w")
            json.dump(json_object, file)
            file.close()
    time.sleep(60)
```
